```python
cols = ['ad_platform', 'ad_type', 'target_gender', 'target_age_group']
for col in cols:
    plt.figure(figsize=(6,3))
    sns.countplot(x=col, data=dataset, palette='viridis')
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(x='ad_platform', hue='target_gender', data=dataset)
plt.title('Ad Platform vs Target Gender')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(x='ad_platform', hue='ad_type', data=dataset)
plt.title('Ad Platform vs Ad Type')
plt.legend(title='Ad Type', loc='upper right', fontsize=8, title_fontsize=9)
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(x='target_age_group', hue='target_gender', data=dataset)
plt.title('Age Group vs Gender')
plt.legend(loc='lower right')
plt.show()
```

```python
from collections import Counter

all_interests = dataset['target_interests'].dropna().str.split(',').sum()
all_interests = [i.strip() for i in all_interests]  # remove spaces
Counter(all_interests).most_common(10)
```

```python
top_interests = pd.Series(Counter(all_interests)).sort_values(ascending=False).head(10)
plt.figure(figsize=(6,3))
top_interests.plot(kind='bar', figsize=(8,4), title='Top 10 Target Interests')
plt.xlabel('Interest')
plt.ylabel('Frequency')
plt.show()
```

```python
dataset[dataset['target_interests'].str.contains('technology', case=False, na=False)]['ad_platform'].value_counts()
```

```python
dataset['ad_platform'].value_counts().plot.pie(
autopct='%1.1f%%', startangle=90, colors=['#1DA1F2','#4267B2'])
plt.title('Ad Platform Share')
plt.ylabel('')
plt.show()
```

```python
dataset['target_gender'].value_counts().plot.pie(
autopct='%1.1f%%', startangle=90, colors=['#FF69B4','#FFD700','#87CEFA'])
plt.title('Target Gender Share')
plt.ylabel('')
plt.show()
```

```python
age_platform = pd.crosstab(dataset['target_age_group'], dataset['ad_platform'])
age_platform.plot(kind='bar', stacked=True, figsize=(6,3), colormap='cool')
plt.title('Age Group vs Platform')
plt.xlabel('Age Group')
plt.ylabel('Ad Count')
plt.legend(title='Platform', loc='lower right')
plt.show()
```

```python
top_campaigns = dataset['campaign_id'].value_counts().head(10)
plt.figure(figsize=(6,3))
sns.barplot(x=top_campaigns.index, y=top_campaigns.values, palette='crest')
plt.title('Top 10 Campaigns by Number of Ads')
plt.xlabel('Campaign ID')
plt.ylabel('Ad Count')
plt.show()
```

```python
dataset['has_technology'] = dataset['target_interests'].str.contains('technology', case=False, na=False)
sns.heatmap(pd.crosstab(dataset['target_gender'], [dataset['ad_platform'], dataset['has_technology']]), annot=True, fmt='d', cmap='YlGnBu')
plt.title('Technology Interest by Gender & Platform')
plt.show()
```

```python
interests = ['technology', 'fashion', 'travel']
for i in interests:
    plt.figure(figsize=(6,3))
    temp = dataset[dataset['target_interests'].str.contains(i, case=False, na=False)]
    count = temp['ad_platform'].value_counts()
    sns.barplot(x=count.index, y=count.values)
    plt.title(f'Ads Targeting "{i}" by Platform')
    plt.show()
```

```python
from wordcloud import WordCloud

all_text = " ".join(dataset['target_interests'].dropna())
wc = WordCloud(width=500, height=250, background_color='white', colormap='viridis').generate(all_text)

plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Target Interests Word Cloud')
plt.show()
```
